```
from keras.models import load_model
from tensorflow import keras
import cv2
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```
kyogoku_model = tf.keras.models.load_model("kyogoku.h5")
ohhara_model = tf.keras.models.load_model("ohhara.h5")
tomii_model = tf.keras.models.load_model("tomii.h5")
yamaoka_model = tf.keras.models.load_model("yamaoka.h5")
kaibara_model = tf.keras.models.load_model("kaibara.h5")
kurita_model = tf.keras.models.load_model("kurita.h5")
tanimura_model = tf.keras.models.load_model("tanimura.h5")
```

```
kyogoku_predict = ["京極万太郎", "その他"]
ohhara_predict = ["大原社主", "その他"]
tomii_predict = ["富井副部長", "その他"]
yamaoka_predict = ["山岡士郎", "その他"]
kaibara_predict = ["海原雄山", "その他"]
kurita_predict = ["その他", "栗田ゆう子"]
tanimura_predict = ["その他", "谷村部長"]
```

```
path = "unknown_001.jpg"
x = 160
y = 90
image = cv2.imread(path, 1)
image = cv2.resize(image, (x, y))
plt.imshow(image)
plt.show()
image = image.reshape(x, y, 3).astype("float64") / 255
```

```
result = []
def load_xception(model, list):
  predict = model.predict(np.array([image]))
  return list[predict[0].argmax()]
```

```
kyogoku = load_xception(kyogoku_model, kyogoku_predict)
result.append(kyogoku)
```

```
ohhara = load_xception(ohhara_model, ohhara_predict)
result.append(ohhara)
```

```
tomii = load_xception(tomii_model, tomii_predict)
result.append(tomii)
```

```
yamaoka = load_xception(yamaoka_model, yamaoka_predict)
result.append(yamaoka)
```

```
kaibara = load_xception(kaibara_model, kaibara_predict)
result.append(kaibara)
```

```
kurita = load_xception(kurita_model, kurita_predict)
result.append(kurita)
```

```
tanimura = load_xception(tanimura_model, tanimura_predict)
result.append(tanimura)
```

```
total = 0
for key, value in enumerate(result):
  if value != "その他":
    print(value)
  if value == "その他":
    total += 1
if total == len(result):
  print("その他")
```
