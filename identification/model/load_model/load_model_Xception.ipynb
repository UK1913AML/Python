```
from keras.models import load_model
from tensorflow import keras
import cv2
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```
path = "unknown_001.jpg"
character = ["大原社主", "山岡士郎", "京極万太郎", "栗田ゆう子", "富井副部長", "海原雄山", "谷村部長"]
x = 160
y = 90
image = cv2.imread(path, 1)
image = cv2.resize(image, (x, y))
plt.imshow(image)
plt.show()
image = image.reshape(x, y, 3).astype("float64") / 255
```

```
model = tf.keras.models.load_model("identification_xception.h5")
predict = model.predict(np.array([image]))
result = predict[0]
for key, value in enumerate(result):
  print(f"{character[key]}: {'{:.2%}'.format(value)}")
print()
print(character[result.argmax()])
```
